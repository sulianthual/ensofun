## ENSO fun
This jupyter notebook is about building a El Nino forecast model from GODAS ocean data reanalysis. We first process and analyze data. Then we perform PCA to determine the leading modes of variability. 

```python
# imports
import numpy as np
import xarray
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# open console
%qtconsole
```

### Part I: Process Observations

The Data is from the GODAS reanalysis: [link](https://psl.noaa.gov/data/gridded/data.godas.html)

We are interested in the followed fields in the equatorial Pacific, that are relevant to predict El Nino: 
* SST anomalies


To obtain these, hereafter we will:
1. Download potential temperature files with global and monthly coverage (pottmp.XX.nc)
  * The files (very large) are already downloaded manually to subfolder ./data
2. Restrict to equatorial Pacific domain: 120E-80W and 10S-10N
3. Compute Fields:
  * SST as potential temperature at first level (5m depth)
4. Remove Climatologies and Trends
5. Compute land areas (with bad values)


#### * Preliminary Analysis:
We analyse first data file for a feel of the data structure

```python
# Analysis of first data file: 
data=xarray.open_dataset('./data/pottmp.1980.nc')

# slice equatorial Pacific domain
# lon_min,lon_max,lat_min,lat_max=120,280,-10,10# restrained Pacific
lon_min,lon_max,lat_min,lat_max=120,280,-30,30# extended Pacific
data=data.sel(lon=slice(lon_min,lon_max),lat=slice(lat_min,lat_max))

# Record axis lon,lat (common to all files)
lon=data["lon"].values# (deg lon) np.ndarray
lat=data["lat"].values# (deg lat)
nx=len(lon)
ny=len(lat)

# Close
data.close()
data
```

#### * Read SST Data:
We read SST through all restart files and concatenate data

```python
def getsst(filename,varname,lon_min,lon_max,lat_min,lat_max):
    '''
    
    Return SST and time over equatorial Pacific for one data file
    from jupyter notebook: ensofun
    
    '''
    # read netcdf variable
    data=xarray.open_dataset(filename)
    # slice equatorial Pacific domain (120E-80W, 10S-10N)
    # slice first vertical level (5m)
    data=data.sel(lon=slice(lon_min,lon_max),lat=slice(lat_min,lat_max),level=5)
    # extract sst and time
    sst=data[varname].values
    time=data["time"].values# (datetime64)
    # convert sst from K to C
    sst=sst-273.15
    # convert time to pseudo axis in years
    dates=pd.DatetimeIndex(time)# convert to date format
    time=np.asarray(dates.year+(dates.month-0.5)/12)# (np.ndarray)  
    data.close()
    return sst,time

# Read SST
sst=np.empty([0,ny,nx])
time=np.empty(0)
for i in range(1980,2020):
    filename='./data/pottmp.'+str(i)+'.nc'
    term1,term2=getsst(filename,'pottmp',lon_min,lon_max,lat_min,lat_max)
    sst=np.concatenate((sst,term1),axis=0)
    time=np.concatenate((time,term2),axis=0)
type(sst),sst.shape,type(time),time.shape
```

#### * Compute Land Map
We detect the longitude and latitudes of land. This will be useful later to omit the land values in the PCA. 

```python
# Detect land map
land=np.empty([ny,nx])
for i in range(nx):
    for j in range(ny):
        foundland=np.any(np.isnan(sst[:,j,i]))
        land[j,i]=foundland#=1 where land, =0 where ocean
        if foundland:# extend bad values to entire timeserie for consistency
            sst[:,j,i]=np.nan
```

```python
def plot_landmap():
    '''
    (Optional): Plot land map
    '''
    plt.contourf(lon,lat,land)
    plt.xlabel('lon (deg E)')
    plt.ylabel('lat (deg N)')
    plt.title('land mask')
plot_landmap()
```

#### * Compute SST Trend, Climatology and anomalies
We compute the SST trend (SSTT), climatology (SSTC) and anomalies (SSTA). The SSTA is more suitable to use for ENSO forecasts and analysis. The total SST reconstructs as: SST= SSTT+SSTC+SSTA

```python
# Compute SST trend
nt=len(time)
sstt=np.empty([nt,ny,nx])
time_trend=time-np.mean(time)
for i in range(nx):
    for j in range(ny):
        term=sst[:,j,i]
        if not np.any(np.isnan(term)):        
            aa,bb=np.polyfit(time_trend,term,1)# first order polynomial fit
            sstt[:,j,i]=aa*time_trend#+bb
```

```python
# Compute SST climatology

nmonths=12
sst=np.reshape(sst,(int(len(time)/nmonths),nmonths,ny,nx))# year,month,lat,lon
nyears=sst.shape[0]
sstclim=np.mean(sst,axis=0)
sstc=np.empty([0,ny,nx])
for k in range(nyears): sstc=np.concatenate((sstc,sstclim),axis=0)
# sstc=np.repeat(sstclim,nyears,axis=0)# replicated sst clim
sst=np.reshape(sst,(nt,ny,nx))# revert to format
```

```python
# Compute SST anomalies
ssta=sst-sstt-sstc
```

#### * (Optional) Plots of SST Climatology and Anomalies
A few plots conclude this part by analysing the climatology and anomalies of SST. We retrieve a consistent SST climatology, with the development of the eastern Pacific cold tongue in boreal winter. 

Meanwhile, the longitude-time hovmollers show the El Nino events over the observation period. The El Nino events show increased SSTA anomalies in the central to eastern Pacific. The major El Nino events with maximal SST anomalies are over 1982/83, 1997/1998 and 2015/2016.

```python
# (Optional) Plot SST Climatology maps

def plot_sstclim():
    '''
    Plot Maps of SST climatology
    '''
    print('Range of SST climatology?:'+str(round(np.nanmin(sstclim),2))+' to '+str(round(np.nanmax(sstclim),2))+' C' )
    fig=plt.figure(1,figsize=(10,10))
    fig.clf()
    fig.suptitle('SST Climatology')
    fig.subplots_adjust(hspace=1,wspace=1)
    mlist=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
    levels=np.linspace(18,31,14)
    for i in range(12): 
        plt.subplot(4,3,i+1)
        plt.contourf(lon,lat,sstclim[i],levels)
        plt.title(mlist[i])
        plt.xlabel('lon (deg E)')
        plt.ylabel('lat (deg N)')
        cbar=plt.colorbar()
        cbar.ax.set_ylabel('deg C')
plot_sstclim()
```

```python
def plot_sstssta():
    '''
    Plot lon-time Hovmollers of SST, SSTA, SSTC (average 5S-5N)
    (total, anomalies, climatology)
    '''
    jmax=np.where(lat>5)[0][0]
    jmin=np.where(lat<-5)[0][-1]
    sst_xt=np.nanmean(sst[:,jmin:jmax,:],axis=1)
    ssta_xt=np.nanmean(ssta[:,jmin:jmax,:],axis=1)
    sstc_xt=np.nanmean(sstc[:,jmin:jmax,:],axis=1)
    sstt_xt=np.nanmean(sstt[:,jmin:jmax,:],axis=1)
    print('Range of SST?:'+str(round(np.nanmin(sst_xt),2))+' to '+str(round(np.nanmax(sst_xt),2))+' C' )
    print('Range of SSTA?:'+str(round(np.nanmin(ssta_xt),2))+' to '+str(round(np.nanmax(ssta_xt),2))+' C' )
    print('Range of SSTC?:'+str(round(np.nanmin(sstc_xt),2))+' to '+str(round(np.nanmax(sstc_xt),2))+' C' )
    print('Range of SSTT?:'+str(round(np.nanmin(sstt_xt),2))+' to '+str(round(np.nanmax(sstt_xt),2))+' C' )

    fig=plt.figure(2,figsize=(10,10))
    fig.subplots_adjust(hspace=1,wspace=1)
    fig.suptitle('SST: Total, Anomalies, Climatology and Trend (average 5S-5N)')

    plt.subplot(1,4,1)
    levels=np.linspace(21,31,11)
#     plt.contourf(lon,time,sst_xt,levels)
    plt.contourf(lon,time,sst_xt)
    plt.title('SST')
    plt.xlabel('lon (deg E)')
    plt.ylabel('time (years)')
    cbar=plt.colorbar()
    cbar.ax.set_ylabel('deg C')

    plt.subplot(1,4,2)
    levels=np.linspace(-6,6,21)
#     plt.contourf(lon,time,ssta_xt,levels)
    plt.contourf(lon,time,ssta_xt)
    plt.title('SSTA')
    plt.xlabel('lon (deg E)')
    plt.ylabel('time (years)')
    cbar=plt.colorbar()
    cbar.ax.set_ylabel('deg C')

    plt.subplot(1,4,3)
    levels=np.linspace(21,31,11)
#     plt.contourf(lon,time,sstc_xt,levels)
    plt.contourf(lon,time,sstc_xt)
    plt.title('SSTC')
    plt.xlabel('lon (deg E)')
    plt.ylabel('time (years)')
    cbar=plt.colorbar()
    cbar.ax.set_ylabel('deg C')    

    plt.subplot(1,4,4)
    levels=np.linspace(21,31,11)
#     plt.contourf(lon,time,sstt_xt,levels)
    plt.contourf(lon,time,sstt_xt)
    plt.title('SSTT')
    plt.xlabel('lon (deg E)')
    plt.ylabel('time (years)')
    cbar=plt.colorbar()
    cbar.ax.set_ylabel('deg C')  
    
plot_sstssta()
```

### Part II: Principal Component Analysis
We perform a Principal Component Analysis (PCA) of the fields. For this the scikit-learn suit is greatly helpful to streamline most steps. The PCA identifies the main directions among which the data varies (that maximize covariance). 

The method is as follows:
1. Reshape input data
   * Here for a time-PCA we reshape SST(time,lat,lon) to the format SST(time,space) 
2. Manage mask values
   * We ignore mask values (that represent land) in the PCA 
2. Standardize the input data (feature scaling)
   * This simply ensures better results even if input data holds large/small values
3. Compute PCs
4. Analyze PCs amd EOFs


The PCA reveals that a few leading modes account for most of the explained variability (>70%). Those leading modes will be retained to perform ENSO forecasts.

```python
# Prepare Input Data (SSTA)
X=np.reshape(ssta,(nt,ny*nx))

# Restrict input data to ocean (exclude land points)
Xland=np.reshape(land,(ny*nx))        
index_ocean=np.where(Xland==0)[0]
X=X[:,index_ocean]
```

* Manage Mask Points:

We Manage mask points (land) by setting them to zero values


```python
# Standardize data (Optional)
if False:
    # Standardize Data
    from sklearn.preprocessing import StandardScaler

    # Generate scaler operator
    scaler=StandardScaler()

    # Standardize
    # X=scaler.fit_transform(X)
    scaler_X=scaler.fit(X)
    X=scaler_X.transform(X)

    # Check mean,std=0,1
    print(X.mean(),X.std())
```

```python
from sklearn.decomposition import PCA
# Generate PCA operator
pca=PCA()
pca.fit(X)
```

#### (Optional) Plot explained variance of first PCs
We plot the first PCAs of SSTA. The first and second PC explain most variance. 

```python
def plot_sstaeofs(npca):
    '''
    Plot the explained variance of first PCAs for SSTA
    '''
    fig=plt.figure(2,figsize=(5,5))
    plt.plot(pca.explained_variance_ratio_[0:npca]*100)
    plt.plot(pca.explained_variance_ratio_[0:npca]*100,'ro')
    plt.title('%variance explained')
    plt.grid()
plot_sstaeofs(5)
```

```python
# Obtain the first PCs and EOFs

npc=2# PCs cutoff
varexpl=pca.explained_variance_ratio_[:npc]*100
pcs=pca.transform(X)[:,:npc]
eofs_ocean=pca.components_[:npc,:]# ocean only
#
eofs=np.empty([npc,ny*nx])+np.nan# full maps
eofs[:,index_ocean]=eofs_ocean

eofs=np.reshape(eofs,(npc,ny,nx))
pcs.shape,eofs.shape
```

#### (Optional) Plot the first PCs and EOFs
We plot the first PCAs and EOFs of SSTA. The first mode (PC1,EOF1) is representative of the El Nino signal: we easily retrieve the major El Nino events on PC1 (1982/83, 1997/1998, 2015/2016). The associated EOF shows the warming in the central to eastern Pacific that is characteristic of El Nino.

The second mode (PC2,EOF2) captures localized warming in the far eastern Pacific, and its intensity varies from an El Nino event to the other. It highlights the diversity of El Nino: in fact, each El Nino event is somewhat unique in terms of structure (as well as timing, intensity,etc...). For some events (called Eastern Pacific events) the warming reaches the far eastern Pacific while for others (called Central Pacific events) it remains in the central Pacific. To reconstruct the SSTs one must sum up the contributions from the first and second mode (as well as the remaining ones). 

```python
# Optional: plot first PCs and EOFs
def plot_pcseofs():
    '''
    Plot the first PCs and EOFs
    '''
    print('Range of EOFs?:'+str(round(np.nanmin(eofs),2))+' to '+str(round(np.nanmax(eofs),2)))
    print('Range of PCs?:'+str(round(np.nanmin(pcs),2))+' to '+str(round(np.nanmax(pcs),2)))

    fig=plt.figure(3,figsize=(10,10))
    fig.subplots_adjust(hspace=0.5,wspace=0.4)

    plt.subplot(2,2,1)
    plt.plot(time,pcs[:,0])
    plt.xlabel('time (years)')
    plt.title('PC1 ('+str(round(varexpl[0],2))+'%)')

    plt.subplot(2,2,2)
    plt.plot(time,pcs[:,1])
    plt.xlabel('time (years)')
    plt.title('PC2 ('+str(round(varexpl[1],2))+'%)')

    plt.subplot(2,2,3)
    plt.contourf(lon,lat,eofs[0,:,:])
    plt.title('EOF1')
    plt.xlabel('lon (deg E)')
    plt.ylabel('lat (deg N)')
    cbar=plt.colorbar()
    cbar.ax.set_ylabel('deg C')

    plt.subplot(2,2,4)
    plt.contourf(lon,lat,eofs[1,:,:])
    plt.title('EOF2')
    plt.xlabel('lon (deg E)')
    plt.ylabel('lat (deg N)')
    cbar=plt.colorbar()
    cbar.ax.set_ylabel('deg C')
plot_pcseofs()
```

The Model is of the form:
\begin{align}
dx/dt=Bx+\xi
\end{align}

where $x$ is the state of the system, $B$ the evolution operator and $\xi$ is stationary white noise.
